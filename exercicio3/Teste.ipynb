### Melhor dia pra se lançar uma promoção com base no dia com a maior concentração de clientes naquela região.

Para realizar essa análise eu optei por dividir os dados em regiões. Para isso utilizei uma biblioteca chamada geohash que divide os dados em "loteamentos" e desta forma é possível ter uma noção de proximidade com base nestas hashs. Elas também pode ser revertida para latitude e longitude.
Após isso, essa nova coluna foi inserida na tabela para facilitar futuras pesquisas por região.

```python
def get_geohash(lat, long):
    ret = []
    for i in range(lat.shape[0]):
        ret.append(geohash.encode(lat.iloc[i], long.iloc[i], precision=5))
    
    return ret

data["geohash"] = get_geohash(data.latitude, data.longitude)
data.head()
```

Para facilitar criei uma função que dado o local ela mostra um gráfico de barras mostrando o volume de visita aos locais por dia da semana. É complicado fazer uma investigação sobre esses resultados de modo mais detalhado, pois eles irão variar de acordo com a posição geográfica escolhida. Mas, como exemplo utilizei o marco zero como referência. Podemos perceber que terça-feira foi é o dia que tem um maior volume de dados. Logo, caso a rede varejista desejasse fazer uma promoção para seus clientes que estivessem no marco zero, os dados sugerem que terça-feira seria o melhor dia, pois iria atingir um volume maior de cliente

```python
days = ["Sunday", "Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday"]
def get_best_day_push(lat, long, precision=5):
    geo_hash = geohash.encode(lat, long, precision=precision)
    return data[data.geohash.apply(lambda x: x.startswith(geo_hash))].day_of_week.value_counts()[days].plot.bar()

# Latitude e Longitude do marco Mero
get_best_day_push(-8.063000, -34.871811)
```

 ### Qual melhor horário para se dar um push de uma propaganda? De acordo com o local desejado e o dia da semana.
 
O melhor horário será aquela que há uma maior concentração de cliente na área. Como explicado no ítem anterior os resultados vão variar a depender dos parâmetros utilizados (neste caso a latitude, longitude e o dia da semana). Usando os dados fornecidos irei estimar qual seria o melhor horário para se colocar uma propaganda caso o local deseja fosse o Marco Zero. Escolhi a terça-feira como exemplo, pois ela foi a que teve o maior volume de dados da etapa anterior.

Com este gráfico podemos observar que três momentos de pico: entre as 9 e 10 horas da manhã, ao meio dia e as 19 horas. Com essas informações a rede varejista pode lançar uma propaganda ou uma promoção nos momentos mais efetivos.

```python
#data.user.value_counts()[data.user.value_counts().values > 10]
#data[data.user == "User_3822"]User_5838, User_2285

def get_best_horario(lat, long, week_day, precision=5):
    geo_hash = geohash.encode(lat, long, precision=precision)
    data[(data.geohash.apply(lambda x: x.startswith(geo_hash))) & (data.day_of_week == week_day)].hour.value_counts().sort_index().plot.bar()

data.head()

get_best_horario(-8.063000, -34.871811, "Tuesday")
```
